```python
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
import regex as re
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.sql import text
from sqlalchemy.ext.declarative import declarative_base
from pathlib import Path

Base = declarative_base()
```

```python
class Asthma(Base):
    __tablename__ = 'us_chronic_asthma'
    extend_existing=True
    Index = Column(Integer, primary_key=True, autoincrement=True)
    YearStart = Column(Integer)
    YearEnd = Column(Integer)
    LocationAbbr = Column(String)
    LocationDesc = Column(String)
    DataSource = Column(String)
    Topic = Column(String)
    Question = Column(String)
    # Response = Column(String, default=None)
    # DataValueUnit = Column(String, default=None)
    DataValueType = Column(String)
    DataValue = Column(Float, default=None)
    DataValueAlt = Column(Float, default=None)
    # DataValueFootnoteSymbol = Column(String, default=None)
    # DatavalueFootnote = Column(String, default=None)
    LowConfidenceLimit = Column(Float, default=None)
    HighConfidenceLimit = Column(Float, default=None)
    StratificationCategory1 = Column(String)
    Stratification1 = Column(String)
    # StratificationCategory2 = Column(String, default=None)
    # Stratification2 = Column(String, default=None)
    # StratificationCategory3 = Column(String, default=None)
    # Stratification3 = Column(String, default=None)
    GeoLocation = Column(String)
    # ResponseID = Column(Integer, default=None)
    LocationID = Column(Integer)
    TopicID = Column(String)
    QuestionID = Column(String)
    DataValueTypeID = Column(String)
    StratificationCategoryID1 = Column(String)
    StratificationID1 = Column(String)
    # StratificationCategoryID2 = Column(String, default=None)
    # StratificationID2 = Column(String, default=None)
    # StratificationCategoryID3 = Column(String, default=None)
    # StratificationID3 = Column(String, default=None)
```

```python
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker

Session = sessionmaker(bind=engine)
session = Session()
```

```python
data_to_import[2]
```

```python
for row in data_to_import:
    print(row['Question'])  
    try:
        existing_entry = session.query(Asthma).filter(Asthma.Question == row['Question']).first()
    except Exception as e:
        print(f"Error querying for Question: {row['Question']}, Error: {e}")
        continue  

    if existing_entry:
        
        continue  
        
    else:
        new_entry = Asthma(
            YearStart=row['YearStart'],
            YearEnd=row['YearEnd'],
            LocationAbbr=row['LocationAbbr'],
            LocationDesc=row['LocationDesc'],
            DataSource=row['DataSource'],
            Topic=row['Topic'],
            Question=row['Question'],
            DataValueType=row['DataValueType'],
            DataValue=row['DataValue'],
            DataValueAlt=row['DataValueAlt'],
            LowConfidenceLimit=row['LowConfidenceLimit'],
            HighConfidenceLimit=row['HighConfidenceLimit'],
            StratificationCategory1=row['StratificationCategory1'],
            Stratification1=row['Stratification1'],
            GeoLocation=row['GeoLocation'],
            LocationID=row['LocationID'],
            TopicID=row['TopicID'],
            QuestionID=row['QuestionID'],
            DataValueTypeID=row['DataValueTypeID'],
            StratificationCategoryID1=row['StratificationCategoryID1'],
            StratificationID1=row['StratificationID1']
        )
        session.add(new_entry)

session.commit()
```
